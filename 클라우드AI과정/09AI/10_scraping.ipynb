```python
import urllib.request
import re

url = "https://finance.naver.com/item/main.nhn?code=017670"
html = urllib.request.urlopen(url)
html_contents = str(html.read().decode("ms949"))

#첫번째 html 패턴
#(<dl class ="blind">)
#(<dd>sdkfjsldjfk<\dd>)
#(<\dl>)
stock_results= re.findall("(\<dl class=\"blind\"\>)([\s\S]+?)(\<\/dl\>)", html_contents)
stock_info=stock_results[0]
stock_index= stock_info[1]
#
print(stock_info)
```

```python
print(stock_index)
```

```python
#주식 정보만 추출
index_list = re.findall("(\<dd\>)([\s\S]+?)(\<\/dd\>)", stock_index) # <- 패턴 추가

for index in index_list :
    print(index[1])
```

```python
from bs4 import BeautifulSoup
import requests

maximum =0
page =1

url = "https://finance.naver.com/news/market_notice.nhn"

response = requests.get(url)
source = response.text
soup = BeautifulSoup(source, "html.parser")


# <a href="/news/market_notice.nhn?&amp;page=4593">
last_page_a = soup.find("td", {"class":"pgRR"})
value = last_page_a.find("a").get("href")
maximum = value.split("=")[1]

print("총" +str(maximum)+"개의 페이지가 확인됨" )


```

```python
whole_source = ""

for page_number in range(1,2):
    url = "https://finance.naver.com/news/market_notice.nhn?&page=" + str(page_number)
    response = requests.get(url)
    whole_source = whole_source +response.text

#contentarea_left > div.boardList2 > table > tbody > tr:nth-child(1) > td.wdate



soup = BeautifulSoup(whole_source, "html.parser")
find_title = soup.select("#contentarea_left > div.boardList2 > table > tr > td.publicSubject > a")
find_date = soup.select("#contentarea_left > div.boardList2 > table > tr > td.wdate")

for idx in range(len(find_title)):
    print(find_title[idx].text,find_date[idx].text) # 날짜 정보도 추가 출력
```
