```python
from pandas_datareader import data
import datetime
import yfinance as yf
yf.pdr_override()
```

```python
start_data = '2010-01-01'
name = '034730.KS'
df= data.get_data_yahoo(name, start_data)
```

```python
df.head(3)
```

```python
df.tail(5)
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df['Close'].plot(figsize= (12,6), grid=True)
```

```python
df_trunc = df[:'2019-12-31']
df_trunc.head(3)
df_trunc.tail(3)

```

```python
new_df = pd.DataFrame({'ds': df_trunc.index, 'y': df_trunc['Close']})
new_df.reset_index(inplace= True)
del new_df['Date']
new_df.head(3)
```

```python
from fbprophet import Prophet

m=Prophet()
m.fit(new_df)
```

```python
future =m.make_future_dataframe(periods= 365*2)
future.tail(3)
```

```python
forecast = m.predict(future)
forecast[['ds','yhat','yhat_lower','yhat_upper']].tail(3)
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.index, df['Close'], label = 'real')
plt.plot(forecast['ds'], forecast['yhat'],label= 'forecast')
plt.grid()
plt.legend()
plt.show()
```
