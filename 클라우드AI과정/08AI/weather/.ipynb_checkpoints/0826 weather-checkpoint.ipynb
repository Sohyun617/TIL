```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

```python
df = pd.read_csv('temp10years.csv', encoding= 'utf-8')
```

```python
df.info()
```

```python
df.head()
```

```python
train_year = (df['연']<= 2015)
test_year = (df['연']>= 2016)
interval = 6
```

```python
#과거 6일의 데이터를 기반으로 해서 학습할 데이터를 생성
def make_data(data):
    #  학습데이터 
    x=[]
    
    # 결과 데이터
    y =[]
    
    temps = list(data["기온"])
    
    for i in range(len(temps)):
        if i < interval : continue
        y.append(temps[i])
        xa =[]
        for p in range(interval):
            d = i+p -interval
            xa.append(temps[d])

        x.append(xa)
        print(xa)
    
    return (x,y)
```

```python
# train, test 데이터 준비
x_train,y_train = make_data(df[train_year])
```

```python
x_test, y_test = make_data(df[test_year])
```

```python
# linear regression 알고리즘 사용
model = LinearRegression(normalize= True)

#학습하기
model.fit(x_train, y_train)

#검증, 예측하기
y_pred = model.predict(x_test)
```

```python
# 결과를 그래프로 표시
plt.figure(figsize= (10,6), dpi = 100)
plt.plot(y_test, c = 'r')
plt.plot(y_pred,c='b')
# plt.savefig("weather-temp.png")
plt.show()
```

```python
#오차 출력
print(y_pred - y_test)
diff_y = abs(y_pred - y_test)
print('평균', sum(diff_y)/len(diff_y))
print('최대오차', max(diff_y))
```
